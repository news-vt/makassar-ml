```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_toy(t: int, w: int, d: int = 1) -> np.ndarray:
    assert d >= 1
    assert t > 0
    assert w > 0

    # Creates matrix of size DxT
    # with repeated rows for time series.
    x = np.arange(t, dtype=np.float64)
    x = np.tile(x, (d, 1))
    for i in range(d):
        s = 0
        for j in range(d):
            if j != i:
                s += np.sin(2*np.pi*(j+1)*x[j]/w)
        x[i] = np.sin(2*np.pi*(i+1)*x[i]/w) + (1./(d-1)) * s

    return x
```

```python
x = make_toy(64, 64, 6)
for i in range(x.shape[0]):
    plt.figure()
    plt.stem(x[i])
```
